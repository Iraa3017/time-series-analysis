```python
import datetime as datetime
```

now = datetime.datetime.now() is a method that returns the current date and time as a datetime object. This object has various attributes, including year, month, and day.

When you write datetime.datetime.now(), you're calling the now() method, which gives you a datetime object.
Once you have that object, you can access its attributes like day, month, etc.

datetime.datetime.now.day() is trying to call day as if it were a method (which it's not), resulting in an error.
You first call now() to get the current datetime object, then access the day attributes

```python
now1=datetime.datetime.now()  # Current date and time
#now1 var is an object now and it will have attributes
#now() is the method
```
